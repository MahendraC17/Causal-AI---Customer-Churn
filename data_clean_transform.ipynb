```python
import pandas as pd
import numpy as np
```

```python
df_dem = pd.read_csv('Telco Customer Churn data\Telco_customer_churn_demographics.csv')
df_loc = pd.read_csv('Telco Customer Churn data\Telco_customer_churn_location.csv')
df_pop = pd.read_csv('Telco Customer Churn data\Telco_customer_churn_population.csv')
df_ser = pd.read_csv('Telco Customer Churn data\Telco_customer_churn_services.csv')
df_sts = pd.read_csv('Telco Customer Churn data\Telco_customer_churn_status.csv')
```

```python
# Cleaning string features
for i in [df_dem, df_loc, df_pop, df_ser, df_sts]:
    i.columns = i.columns.str.replace(' ', '_').str.lower().str.strip()
    for j in i.columns:
        if str(i[j].dtypes) == 'object':
            i[j] = i[j].str.replace(' ', '_')
            i[j] = i[j].str.replace('-', '')
```

```python
df_list = [df_dem, df_loc, df_pop, df_ser, df_sts]
df_type = ['dem', 'loc', 'pop', 'ser', 'sts']
```

```python
# Function to get a list of features which has 'yes' and 'no' category of data from all of the df
def get_cat_yesno(df_list, df_type):
    dict = {}
    num = 0
    for i in df_list:
        list = []
        for j in i.columns:
            if i[j].unique().tolist() == ['No', 'Yes'] or i[j].unique().tolist() == ['Yes', 'No']:
                list.append(j)
        dict[df_type[num]] = list
        num += 1
    return dict

df_cat_yesno = get_cat_yesno(df_list, df_type)
```

```python
# Function to transform the 'yes' and 'no' category features to binary feartures
def df_cat_transform(cat_data, list):
    num = 0
    for i in cat_data.values():
        for j in i:
            list[num][j] = list[num][j].replace({'Yes': 1, 'No': 0})
        num += 1

df_cat_transform(df_cat_yesno, df_list)
```

```python
df_dem.drop(['count'], axis = 1, inplace=True)
df_dem.head()
```

```python
df_loc.drop(['count'], axis = 1, inplace=True)
df_loc.head()
```

```python
df_pop.drop(['id'], axis = 1, inplace=True)
df_pop.head()
```

```python
df_ser.replace({None: np.nan, '': np.nan}, inplace=True)
df_ser.head()
```

```python
df_sts.replace({None: np.nan, '':np.nan}, inplace=True)
df_sts.head()
```

```python
df_dem_loc = pd.merge( df_dem, df_loc, on = 'customer_id')
df_dem_loc_pop = pd.merge(df_dem_loc, df_pop, on = 'zip_code')
df_dem_loc_pop_ser = pd.merge(df_dem_loc_pop, df_ser, on = 'customer_id')
df_dem_loc_pop_ser_sts = pd.merge(df_dem_loc_pop_ser, df_sts, on = 'customer_id')
df_dem_loc_pop_ser_sts.head()
```

```python
df = df_dem_loc_pop_ser_sts
df.shape
```
